# Neptunian desert plots — configuration notebook

This notebook is a light front-end for the plotting code in `nep_des_core.py`.

1. Edit and run the **USER CONFIGURATION** cell below.
2. Run the final cell to push the configuration into the core module and generate the three figures:
   - Approximate P–R landscape
   - P–R + P–ρ density two-panel plot
   - DR25 exact-boundaries plot.


```
# ===================================================================
# USER CONFIGURATION
# ===================================================================
# This is the main configuration cell. The user is expected to edit:
#
#   1–4: required for ALL plots
#        - P–R approximate landscape
#        - P–R + P–ρ density two–panel plot
#        - DR25 "exact boundaries" plot
#
#   5–6: only affect the P–ρ density trend (right panel of the
#        two–panel figure) and its Monte Carlo statistics.
#
#   7:   optional tag appended to all output filenames.
#
# All other cells below are "library code" and normally do not need to
# be modified by the user.
# ===================================================================

# -------------------------------------------------------------------
# 1) Catalogue used for the landscape and two–panel plots
#    -> Affects: P–R approximate landscape, P–R and P–ρ density plot
#       (NOT the DR25 "exact boundaries" plot, which uses its own file)
# -------------------------------------------------------------------
# Options:
#   - "NEA"          : NASA Exoplanet Archive
#   - "Exoplanet.eu" : exoplanet.eu catalogue
catalog = "NEA"


# -------------------------------------------------------------------
# 2) Neptunian landscape boundaries and labels
#    -> Affects: ALL P–R panels
#       (approximate landscape, density plot left panel, exact/DR25)
# -------------------------------------------------------------------
# Which boundaries to draw in the P–R plots (approx + density + exact).
# Comma–separated list of keys:
#   - "desert" -> draw the Castro-González et al. (2024) Neptunian desert
#   - "ridge"  -> draw the Neptunian ridge (vertical line / oval)
show_boundaries = "desert,ridge"

# Which region labels to display (comma–separated subset of):
#   "desert", "ridge", "savanna"
show_labels = "desert,ridge,savanna"

# Whether to overplot the classical Neptunian desert limits from
# Mazeh et al. (2016) as dotted lines in all P–R panels.
show_mazeh_boundaries = False


# -------------------------------------------------------------------
# 3) User planets: single definition used by ALL plots
#    -> Affects: ALL plots that show individual systems
#       (P–R approximate, density plot, exact/DR25, P–ρ density panel)
# -------------------------------------------------------------------
# Expected fields for each planet dictionary:
#   name             : str, planet name (used in legends/labels)
#   P_days           : float, orbital period in days
#   R_Rearth         : float, radius in Earth radii (R⊕)
#   R_err_up/down    : float, asymmetric 1σ errors on the radius (R⊕)
#   M_Mearth         : float, optional mass in Earth masses (M⊕)
#   M_err_up/down    : float, optional 1σ errors on the mass (M⊕)
#   rho_rhoearth     : float, optional density in units of ρ⊕
#   rho_err_up/down  : float, optional 1σ errors on the density (ρ⊕)
#   M_upper_3sigma   : bool (optional), if True, interpret mass as a
#                      3σ upper limit
#   rho_upper_3sigma : bool (optional), if True, interpret density as a
#                      3σ upper limit
#   color            : marker color in all plots
#   dx_label,dy_label: linear offsets for the label position in P–R plots
#
# Notes:
# - If M and R are provided, ρ is computed automatically with error
#   propagation.
# - If rho_rhoearth is provided, it is used directly (M and R become
#   optional for the density plot).
# - For the density trend plot:
#   * If neither (M,R) nor rho_rhoearth are available, the planet is not
#     plotted and a WARNING is issued.
#   * Upper limits (M_upper_3sigma / rho_upper_3sigma = True) are drawn
#     with downward arrows in the P–ρ panel.

user_planets = [
    # ---------------------------------------------------------------
    # TOI-5005 b
    # Source: Castro-González et al. (2024, A&A) / NASA Exoplanet Archive
    # P = 6.3 d, Rp = 6.25 ± 0.24 R⊕, Mp = 32.7 ± 5.9 M⊕
    # ---------------------------------------------------------------
    dict(
        name="TOI-5005 b",
        P_days=6.3,
        R_Rearth=6.25,
        R_err_up=0.24,
        R_err_down=0.24,
        M_Mearth=32.7,
        M_err_up=5.9,
        M_err_down=5.9,
        color="#ffcc00",
        dx_label=0.0,
        dy_label=0.0,
    ),

    # ---------------------------------------------------------------
    # TOI-849 b
    # Source: Open Exoplanet Catalogue / Exoplanet Kyoto / exoplanet.eu
    # P = 0.7655 d, Rp = 3.44(+0.16/-0.11) R⊕, Mp = 39 ± 3 M⊕
    # ---------------------------------------------------------------
    dict(
        name="TOI-849 b",
        P_days=0.7655,
        R_Rearth=3.44,
        R_err_up=0.16,
        R_err_down=0.11,
        M_Mearth=39.0,
        M_err_up=3.0,
        M_err_down=3.0,
        color="#00ccff",
        dx_label=0.0,
        dy_label=0.0,
    ),

    # ---------------------------------------------------------------
    # Example of a 3σ upper limit in mass
    # Drawn as an upper–limit symbol in the P–ρ density panel.
    # ---------------------------------------------------------------
    dict(
        name="Example 3σ UL",
        P_days=3.0,           # orbital period in days
        R_Rearth=5.5,         # radius in Earth radii
        R_err_up=0.20,
        R_err_down=0.20,
        M_Mearth=20.0,        # 3σ upper limit: M < 20 M⊕
        M_err_up=0.0,
        M_err_down=0.0,
        M_upper_3sigma=True,  # treated as a 3σ upper limit
        color="#FF6666",
        dx_label=0.0,
        dy_label=0.0,
    ),
]

# To disable user planets entirely, set:
#   user_planets = []


# -------------------------------------------------------------------
# 4) How to display highlighted (user) planets
#    -> Affects: ALL plots that include user planets
#       (approximate landscape, density plot, exact/DR25 plot)
# -------------------------------------------------------------------
# Options:
#   - "box"    -> inline labels inside the plot, with colored text boxes
#   - "legend" -> classic legend with one symbol per planet
legend_or_box = "legend"


# -------------------------------------------------------------------
# 5) Radius range and density–plot configuration
#    -> Affects:
#       - R_band_min / R_band_max:
#           * P–R approximate landscape (shaded Neptunian range)
#           * P–R left panel of the density plot
#           * Sample selection for the P–ρ density trend
#       - density_ymax / density_xlim_max:
#           * Only the P–ρ panel (right panel of the density plot)
# -------------------------------------------------------------------
# Radius interval (in Earth radii) used to define the Neptunian range
# and to select planets for the density trend.
R_band_min = 4.5   # R_earth (default)
R_band_max = 7.5   # R_earth (default)

# Maximum density and period range for the P–ρ panel.
# - density_xlim_max: ALSO used to define the savanna sample
#   (P ≤ density_xlim_max) in the density statistics.
# - density_ymax: only affects the vertical axis limits (visualization).
density_ymax     = 3.2   # g cm^-3 (upper y–limit, display only)
density_xlim_max = 23.0  # days (upper x–limit AND sample cut)


# -------------------------------------------------------------------
# 6) Precision cuts for the density trend (P–ρ panel ONLY)
#    -> Affects: catalogue selection for the density trend and all
#       Monte Carlo density statistics (Desert / Ridge / Savanna)
# -------------------------------------------------------------------
# Relative density–uncertainty cut:
#   If max_rel_err_rho_pct > 0.0, we apply:
#       sigma_rho / rho < max_rel_err_rho_pct / 100
#   If max_rel_err_rho_pct <= 0.0, we fall back to the legacy cuts
#   on sigma_R/R and sigma_M/M defined below.
max_rel_err_rho_pct = 33.0   # e.g. 40.0 means 40% relative error in density

# Legacy cuts (only used if max_rel_err_rho_pct <= 0.0).
# These are also expressed as percentages.
max_frac_err_R_pct = 33.0    # max 1σ fractional error in R (in %)
max_frac_err_M_pct = 33.0    # max 1σ fractional error in M (in %)

# Whether to include user planets in the density Monte Carlo statistics.
# -> Affects ONLY the density trend statistics (not the P–R plots).
include_user_planets_in_density_stats = False


# -------------------------------------------------------------------
# 7) Optional tag for output filenames
#    -> Affects: ALL figures saved to the 'output' directory.
# -------------------------------------------------------------------
# If empty: figures are saved with their default names, e.g.:
#   neptunian_landscape_approx.pdf
# If non-empty (e.g. "TOI-5005"), all figure names become:
#   "<base_name>_<plot_tag>.pdf/png/jpg", for example:
#   neptunian_landscape_approx_TOI-5005.pdf
plot_tag = ""  # e.g. "TOI-5005"
```

```
import nep_des_core as nd

# ------------------------------------------------------------------
# Push the current notebook configuration into the core module
# ------------------------------------------------------------------
nd.catalog   = catalog
nd.show_boundaries = show_boundaries
nd.show_labels     = show_labels
nd.show_mazeh_boundaries = show_mazeh_boundaries

nd.user_planets   = user_planets
nd.legend_or_box  = legend_or_box

nd.R_band_min     = R_band_min
nd.R_band_max     = R_band_max
nd.density_ymax   = density_ymax
nd.density_xlim_max = density_xlim_max

nd.max_rel_err_rho_pct = max_rel_err_rho_pct
nd.max_frac_err_R_pct  = max_frac_err_R_pct
nd.max_frac_err_M_pct  = max_frac_err_M_pct
nd.include_user_planets_in_density_stats = include_user_planets_in_density_stats

nd.plot_tag = plot_tag

# ------------------------------------------------------------------
# Recompute catalogue-dependent arrays in the core module
# ------------------------------------------------------------------
nd.df_cat, nd.P_catalog, nd.R_catalog = nd.load_catalog_PR(nd.catalog)
nd.logP_kde, nd.logR_kde = nd.load_frozen_kde_points(nd.catalog)

# ------------------------------------------------------------------
# Generate the three figures
# ------------------------------------------------------------------
nd.plot_neptunian_landscape_approx()
nd.plot_neptunian_landscape_density()
nd.plot_neptunian_landscape_exact()
```
